## The Statistics of the NBA 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df= pd.read_csv("all_seasons.csv")
df

```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['player_weight'], df['player_height'], alpha=0.7)
plt.title('Average of player Weight vs Height')
plt.xlabel('Weight (In Kilos)')
plt.ylabel('Height (In Centimeters)')
plt.grid(True)
plt.show()
```

```python
#What I noticed from the Graph was that on average, the heavier a NBA player is, the taller they are.
#I also learned that the shortest player of the NBA stood at a whopping 160 cm, that's 5'3"!
#I also learned that it is packed mostly between 80 and 120 kilos and 180 to 220 centimeters.
```

```python
df1=df.sort_values(by="draft_year")
plt.figure(figsize=(10, 6))
plt.bar(df1['draft_year'], df1['player_height'], color='skyblue')
plt.title('Draft height over the years')
plt.xlabel('Year')
plt.ylabel('Height (In Centimeters)')
plt.xticks(rotation=90)
plt.show()
```

```python
#From this graph I learned that the shortest draft height was in the year 1963
#I also learned that the tallest draft height was in the year 1993
#I also have seen that the average draft height does not vary much from year to year 1963
```

```python
plt.figure(figsize=(10, 6))  


plt.bar(df['age'], df['player_weight'], width=0.5, color='skyblue', edgecolor='black')

plt.title('Average Weight for Age of NBA Players')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show() #change graph
```

```python
#I learned that majority players are between 20 to 40 years old
##I also learned that NBA players mostly weigh between 80 and 120 kilos
```
